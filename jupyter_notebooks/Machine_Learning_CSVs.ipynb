```python
import pandas as pd
import os

file = pd.read_csv("kepler_data.csv")

planets_df = pd.DataFrame(file)
planets_df
```

```python
planet_column_names = pd.read_csv("planet_columns.csv")
planet_column_names = planet_column_names.df_columns
planet_column_names = planet_column_names.to_list()
planet_column_names
```

```python
#replace df column names with more legible columns
planets_df.columns = planet_column_names
planets_df
```

```python
empty_columns = []
for column in planets_df:
    column_name=column
    if (planets_df[column].isna().sum()/len(planets_df[column]) > 0.20):
        empty_columns.append(column_name)

empty_columns
```

```python
planets_df=planets_df.drop(columns=empty_columns, axis=1)
planets_df
```

```python
# Exoplanet_Archive_Disposition will ultimately be our target
planets_df=planets_df.drop(columns=['Disposition_Using_Kepler_Data','Disposition_Score'], axis=1)
planets_df
```

```python
# drop False_Positive_Flag columns because they are basically giving us the answer we are trying to calculate
planets_df=planets_df.drop(columns=['Not_Transit-Like_False_Positive_Flag','Stellar_Eclipse_False_Positive_Flag', 'Centroid_Offset_False_Positive_Flag', 'Ephemeris_Match_Indicates_Contamination_False_Positive_Flag'], axis=1)
planets_df
```

```python
new_planets_df=planets_df.set_index('KOI_Name')
new_planets_df
```

```python
new_planets_df = new_planets_df.drop(columns=['Orbital_Period_Upper_Unc._[days]','TCE_Planet_Number', 'Orbital_Period_Lower_Unc._[days]', 'Stellar_Effective_Temperature_Lower_Unc._[K]', 'Stellar_Effective_Temperature_Upper_Unc._[K]', 'Stellar_Surface_Gravity_Upper_Unc._[log10(cm/s**2)]', 'Stellar_Surface_Gravity_Lower_Unc._[log10(cm/s**2)]'])
new_planets_df = new_planets_df.drop(columns=['Transit_Epoch_Upper_Unc._[BKJD]', 'Transit_Epoch_Lower_Unc._[BKJD]', 'Stellar_Radius_Upper_Unc._[Solar_radii]', 'Stellar_Radius_Lower_Unc._[Solar_radii]'])
new_planets_df = new_planets_df.drop(columns=['Insolation_Flux_Lower_Unc._[Earth_flux]','Insolation_Flux_Upper_Unc._[Earth_flux]','TCE_Delivery'])
new_planets_df = new_planets_df.drop(columns=['Impact_Parameter_Upper_Unc.','Impact_Parameter_Lower_Unc.'])
new_planets_df = new_planets_df.drop(columns=['Transit_Depth_Upper_Unc._[ppm]','Transit_Depth_Lower_Unc._[ppm]'])
new_planets_df = new_planets_df.drop(columns=['Transit_Duration_Upper_Unc._[hrs]','Transit_Duration_Lower_Unc._[hrs]'])
new_planets_df = new_planets_df.drop(columns=['KepID','Planetary_Radius_Upper_Unc._[Earth_radii]','Planetary_Radius_Lower_Unc._[Earth_radii]'])
new_planets_df
```

```python
new_planets_df = new_planets_df.dropna()
new_planets_df
```

```python
confirmed_False_planets_df = new_planets_df.loc[new_planets_df['Exoplanet_Archive_Disposition']!='CANDIDATE']
confirmed_False_planets_df
```

```python
Candidates_Planets_df = new_planets_df.loc[new_planets_df['Exoplanet_Archive_Disposition']=='CANDIDATE']
Candidates_Planets_df
```

```python
confirmed_False_planets_df.to_csv("confirmed_False_planets.csv")
```

```python
Candidates_Planets_df.to_csv("Candidate_planets.csv")
```
